grdientBoosting VS DecisionTree

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import KFold, cross_val_score
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.tree import DecisionTreeClassifier
```

```python
pipe_gradientBoos = make_pipeline(
    StandardScaler(),
    GradientBoostingClassifier()
)
```

```python
iris = load_iris(as_frame=True)
X = iris.data
y = iris.target
y
y = (y==2).astype('int8')
```

```python
kfold = KFold(n_splits=5, shuffle=True)
cross_val_score(pipe_gradientBoos,X,y, cv=kfold, scoring='f1').mean()
```

```python
pipe_decision = make_pipeline(
    StandardScaler(),
    GradientBoostingClassifier()
)
```

```python
cross_val_score(pipe_decision,X,y, cv=kfold, scoring='f1').mean()
```
