```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def linear_Predict(X, w):
    return np.dot(X, w)

def mse_loss(y_hat, y):
    error = y_hat-y
    return error, np.sum(error*error)/error.shape[0]

def calc_gradient(X, error):
    return np.dot(X.T,  error)

def update_weight(w, lr, gradient):
    return w-(lr*gradient)

def bgd(X, y, w, lr, n_epochs):
    losses = []
    W = []
    for epoch in range(n_epochs):
        y_hat = linear_Predict(X, w)
        error, loss = mse_loss(y_hat, y)
        w0,w1 = w[0][0], w[1][0]
        losses.append(loss)
        W.append([w[0][0], w[1][0]])
        gradient = calc_gradient(X, error)
        w = update_weight(w, lr, gradient)

    return losses, W
```

```python
np.random.seed(7)
n_epochs = 1000
lr = 0.0001
x = np.random.randint(1,50,(10,1))
x0 = np.ones_like(x)
X = np.concatenate((x0,x), axis=1)
y = np.random.randint(0,20,(10,1))
w = np.random.rand(X.shape[1],1)
losses, W = bgd(X, y, w, lr, n_epochs)
losses[-1]
```

```python
epochs = np.arange(n_epochs)
plt.plot(epochs,losses)
plt.show()
```
