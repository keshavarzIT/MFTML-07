```python
import numpy as np
import matplotlib.pyplot as plt

# ŸÖÿ≠ÿØŸàÿØŸá ÿ∂ÿ±ÿß€åÿ® ŸÅÿ±ÿ∂€å
w1 = np.linspace(-3, 3, 200)
w2 = np.linspace(-3, 3, 200)
W1, W2 = np.meshgrid(w1, w2)

# ŸæŸÜÿßŸÑÿ™€å L1 Ÿà L2
L1 = np.abs(W1) + np.abs(W2)    # Lasso
L2 = W1**2 + W2**2              # Ridge

# ŸÜŸÖŸàÿØÿßÿ± ⁄©ÿßŸÜÿ™Ÿàÿ±
plt.figure(figsize=(12,5))

# L1
plt.subplot(1,2,1)
plt.contour(W1, W2, L1, levels=20, cmap='Reds')
plt.title('Lasso (L1) Penalty')
plt.xlabel('w1')
plt.ylabel('w2')
plt.grid(True)

# L2
plt.subplot(1,2,2)
plt.contour(W1, W2, L2, levels=20, cmap='Blues')
plt.title('Ridge (L2) Penalty')
plt.xlabel('w1')
plt.ylabel('w2')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

w = np.linspace(-3, 3, 200)
lambda_ = 1.0


lasso_penalty = lambda_ * np.abs(w)
ridge_penalty = lambda_ * w**2


lasso_grad = lambda_ * np.sign(w)
ridge_grad = 2 * lambda_ * w

plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(w, lasso_penalty, label='Lasso Penalty', color='red')
plt.plot(w, ridge_penalty, label='Ridge Penalty', color='blue')
plt.title('Penalty')
plt.xlabel('w')
plt.ylabel('Penalty')
plt.grid(True)
plt.legend()


plt.subplot(1,2,2)
plt.plot(w, lasso_grad, label='Lasso Gradient', color='red')
plt.plot(w, ridge_grad, label='Ridge Gradient', color='blue')
plt.title('Penalty gradient')
plt.xlabel('w')
plt.ylabel('gradient')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
